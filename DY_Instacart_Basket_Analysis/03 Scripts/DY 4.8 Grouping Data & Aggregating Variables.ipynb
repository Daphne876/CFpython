```python
# Finding the mean

ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean']})
```

## Step 6

```python
# Mean price for each user

user_mean_price = ords_prods_merge.groupby('user_id')['prices'].transform('mean')
```

```python
# Assign labels

ords_prods_merge['spending_flag'] = np.where(user_mean_price >= 10, 'high spender', 'low spender')
```

```python
ords_prods_merge.loc[ords_prods_merge['spending_flag'] == 'low spender', 'spending_flag'] = 'regular spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

## Step 7

```python
# Calculate the median

ords_prods_merge['median_days'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Create order frequency flag

ords_prods_merge['frequency_flag'] = 'Non-frequent customer'  
```

```python
# Assign labels

ords_prods_merge.loc[ords_prods_merge['median_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_days'] > 10) & (ords_prods_merge['median_days'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
